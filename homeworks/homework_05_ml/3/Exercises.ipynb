### Step 3. Загрузите в переменную crime

```python
crime = pd.read_csv('US_Crime_Rates_1960_2014.csv', sep=',')
crime.head()
```

### Step 4. Посмотрите на тип колонок.

```python
crime.dtypes
```

##### Year в типе int64. У pandas есть встроенный тип для работы с Временными рядами (Time Series)

### Step 5. Конвертируйте тип Year в pd.datetime64

```python
crime.Year = pd.to_datetime(crime.Year, format='%Y')
crime.dtypes
```

### Step 6. Установите Year как index нашего датафрейма

```python
crime.set_index('Year', drop=True, inplace=True)
crime.head()
```

### Step 7. Удалите колонку Total

```python
crime = crime.drop(['Total'], axis=1)
crime.head()
```

### Step 8. Сгруппируйте года по декадам и просуммируйте некоторые значения.

#### Осторожнее с Population, суммирование этой колонки - неправильно. Что вместо этого с ней лучше сделать?

```python
# группировка по декадам
crime_decades = crime.groupby(lambda x: x.year // 10)

# среднее по населению
print(f"Decades Population mean:\n{crime_decades.Population.mean().astype(int)}")
# crime_decades.agg({'Population': 'mean'}).astype(int)

# сумма по остальным полям
print(f"\nDecades sum:\n{crime_decades[crime.columns[1:]].sum()}")

# среднее по населению и сумма по остальным полям (способ второй)
crime_decades.agg(dict(zip(crime.columns, ['mean'] + ['sum'] * (len(crime.columns)-1))))
```

### Step 9. Какая самая опасная декада в US?

```python
# группировка по декадам
crime_decades = crime.groupby(lambda x: x.year // 10)

# Total ~= Violent + Property
decade = crime_decades.agg(arg={'Violent': 'sum', 'Property': 'sum'}).sum(axis=1).idxmax()
print(f"Самая опасная декада: {decade}")
```
